Q. Compare and contrast mutable and immutable objects in 	Python with examples.

Ans. Mutable Objects:
Mutable objects are those which can be changed after creation.
Examples: Lists, dictionaries, sets.

  Immutable Objects:
Immutable objects are those which cannot be changed after creation.
Examples: Strings, tuples, integers, float.

# Key Differences:
* Modification:
Mutable: The objects can be modified after the creation.
Immutable: Objects can not be modified after the creation.

* Hashability:
Mutable: Not hashable(cannot be used as keys in dictionaries or elements in sets).
Immutable: Hashable(can be used as keys in dictionaries and elements in sets).

* Memory efficiency:
Mutable: It may consumes more memory as compared to immutable objects due to changes.
Immutable: Typically memory efficient as new objects are not created for changes.

Examples:
Mutable: my_list-[1,2,3]<br>my_list.append(4).
Immutable: my_string - "hello"<br>my_string.upper().

Q. Explain the concept of type casting in Python with examples.

Ans. The process of converting a value from one data type to another data type is called Typecasting.This can be done explicitly using built-in functions.

Examples:

```python
1. String to Integer:
           str_num = "5"
           int_num = int(str_num)
           print(int_num)  # Output: 5

2. Float to Integer:
          num = 2.5
          int_num = int(num)
          print(int_num)  # Output: 2

3. Integer to Float:
          num = 2
          float_num = float(num)
          print(float_num)  # Output: 2.0

4. Integer to String:
          num = 23
          str_num = str(num)
          print(str_num)  # Output: "23"





```

Q. How do conditional statements work in Python? Illustrate with examples.

Ans. Conditional statements are those which describe the relationship between two events based on certain conditions.

```python
Examples:
1.`if`:
       a = 50
       if a > 20:
          print ("The number is greater than 20")  # Output ("The number is greater than 20")

2.`if-else`:
       a = 3
       if a > 5:
          print("a is greater than 5")
       else:
          print("a is not greater than 5")  # Output ("a is not greater than 5")

3.`elif`:
       a = 5
       if a > 5:
         print("a is greater than 5")
       elif a == 5:
         print("a is equal to 5")
       else:
         print("a is less than 5")  # Output ("a is equal to 5")
```

Q. Describe the different types of loops in Python and their use cases with examples.

Ans. Python has two main types of loops: `for` loops and `while` loops.It executes a block of code multiple times based on a condition.
1. For Loop: Used to iterate over a sequence of elements,(like string or set,list,tuple,dictionary).

```python
Example:
        Fruits = ["apple", "banana", "cherry"]
        for Fruit in Fruits:
            print(Fruit)  # Output: ("apple,banana,cherry")
```

2. While Loop: Repeats a block of code as long as a specified condition is true.

```python
#Print numbers from 1 to 9

       count = 1
       while count <= 9:
          print(count)
          count += 1       # Output: ("1,2,3,4,5,6,7,8,9")
```

3. Nested Loops: You can use one loop inside another loop.Useful for multi-dimensional data.

```python
Example:
       for i in range(3):
          for j in range(2):
              print(f"i: {i}, {j}")  # Output: i:0, j:0, i:0, j:0........

```
